#### PLOTTING  
The `plot()` function is used to draw points in graphs.  
It takes two parameters:   
An array rep the **x-axis**   
An array rep the **y-axis**   
The no of points does not matter as long as there is same no in both x and y axes.  
Use `'o'` to plot ponits only without a line.  
If only one axis is used, there are **default x-points** i.e 0,1,2,3...  

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1.2, 4, 5.4, 8])
y = np.array([4, 5, 7.8, 10.4])

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'o')
plt.show()
```

```python
plt.plot(y, "o")
plt.show()
```
