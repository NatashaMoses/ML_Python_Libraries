#### SNS.PAIRPLOT()
This uses scatterplot() for each pairing of the variables.  
And histplot() for the marginal plots along the diagonal.

```python
import seaborn as sns
penguins = sns.load_dataset("penguins")

print(penguins.head())

print(penguins.info())

sns.pairplot(penguins)
```

#### HUE
The points are in 2D but a third dimension can be added by coloring the points according to a 3rd variable.  
This converts the histograms to KDE Kernel Density Estimate.    

```python
sns.pairplot(penguins, hue="species")
```

#### DIAG_KIND
This is used to force the diagonals to histograms, or any other kind.  

```python
sns.pairplot(penguins, hue="species", diag_kind="hist")
```
