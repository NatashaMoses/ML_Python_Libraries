#### SEABORN
This is a Python library for making statistical graphics.  

#### SNS.RELPLOT()
This is used to create scatter plots.  

```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

print(tips)

# Create a visualization
sns.relplot(data=tips, x="total_bill", y="tip")
```

#### HUE
The points are in 2D but a third dimension can be added by coloring the points according to a 3rd variable.  
Eg you can now see the points in reference to the day.  

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day")
```

#### STYLE
Style is used to add a different style to each marker.  

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="day", style="day"
)
```

#### HUE AND STYLE TO REP 4 VARIABLES
To represent 4 variables, change the style and hue for each point.  
In this case if thu blue, fri orange, sat green, sun red  
Also if thu lunch blue circle, but thu dinner blue cross  

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="day", style="time",
)
```

#### HUE CATEGORICAL, HUE NUMERICAL
If hue is **categorical** as above, the default qualitative palette is used.  
If hue is **numerical**, a sequential palette is used.  

```python
sns.relplot(
    data=tips, x="total_bill", y="tip", hue="size",
)
```

#### CHANGE HUE PALETTE
This is to change the palette.  

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip",
    hue="size", palette="ch:r=-.5,l=.75"
)
```

#### SIZE
This is used to change the size of each point according  to a third variable.  

```python
sns.relplot(data=tips, x="total_bill", y="tip", size="size")
```

#### SPECIFY SIZE RANGE
The range of the smallest to the largest dot can be specified.  

```python
sns.relplot(
    data=tips, x="total_bill", y="tip",
    size="size", sizes=(15, 200)
)
```
