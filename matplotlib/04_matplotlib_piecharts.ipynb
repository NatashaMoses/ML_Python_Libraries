#### CREATE PIE CHARTS
Pie charts plot a wedge for every observation.  
Use the `pie()` function to create pie charts.  
It is used in the **Pyplot** submodule.   
By default the plotting of the first wedge starts from the x-axis and moves anti-clockwise.  

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([20, 35, 15, 30])

plt.pie(x)
plt.show() 
```

#### LABELS
Use the `labels` parameter to add labels to the pie chart.  
+ First create an array for labels that has same length.  
For example, if you have 4 points in the pie chart, create an array with 4 points.  
Eg `my_labels = np.array(["Cats", "Dogs", "Ducks", "Geese"])`  
+ Use the keyword argument `labels`.   
Eg `plt.pie(x, labels=my_labels)`  

```python
# ADD LABELS
import matplotlib.pyplot as plt
import numpy as np

x = np.array([20, 35, 15, 30])
my_labels = np.array(["Cats", "Dogs", "Ducks", "Geese"])

plt.pie(x, labels = my_labels)
plt.show() 
```

#### START ANGLE
The default start angle is the **x-axis**.  
Use the `startangle` parameter to change the start angle.  
The default start angle is 0 degrees i.e the x-axis.  
To start the first wedge at y-axis use 90 degrees.  
Eg `plt.pie(x, labels = my_labels, startangle = 90)`  

```python
# START ANGLE
import matplotlib.pyplot as plt
import numpy as np

x = np.array([20, 35, 15, 30])
my_labels = np.array(["Cats", "Dogs", "Ducks", "Geese"])

plt.pie(x, labels = my_labels, startangle=90)
plt.show() 
```

#### EXPLODE
You can explode the wedges.  
This makes the wedge separate from the other wedges.  
Use the `explode` parameter to explode a wedge.  
+ First create an array for explode that has same length.  
For example, if you have 4 points in the pie chart, create an array with 4 points.  
Eg `my_explode = [0.3, 0, 0.1, 0]`  
+ Use the keyword argument `explode`.   
Eg `plt.pie(x, labels=my_labels, explode=my_explode)`  

```python
# EXPLODE A WEDGE
import matplotlib.pyplot as plt
import numpy as np

x = np.array([20, 35, 15, 30])
my_labels = np.array(["Cats", "Dogs", "Ducks", "Geese"])
my_explode = [0.3,0,0.1,0]

plt.pie(x, labels = my_labels, explode=my_explode)
plt.show() 
```

#### SHADOW
Use the `shadow` parameter to add a a shadow to the pie chart.  
Eg `plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)`  

```python
# ADD SHADOW
import matplotlib.pyplot as plt
import numpy as np

x = np.array([20, 35, 15, 30])
my_labels = np.array(["Cats", "Dogs", "Ducks", "Geese"])
my_explode = [0.3,0,0.1,0]

plt.pie(x, labels = my_labels, explode=my_explode, shadow=True)
plt.show() 
```

#### COLORS
Use the `colors` parameter to specify the color of a wedge.  
+ First create an array for colors that has same length.  
For example, if you have 4 points in the pie chart, create an array with 4 points.  
Eg `my_colors = ["magnetta", "crimson", "b", "#4CAF50"]`  
You can use hexadecimal color values, any of the 140 supported color names,  or the supported shortcuts.  
+ Use the keyword argument `colors`.   
Eg `plt.pie(x, labels=my_labels, explode=my_explode, colors=my_colors)`  

```python
# COLOR FOR EACH WEDGE
import matplotlib.pyplot as plt
import numpy as np

x = np.array([20, 35, 15, 30])
my_labels = np.array(["Cats", "Dogs", "Ducks", "Geese"])
my_explode = [0.3,0,0.1,0]
my_colors = ["m", "crimson", "b", "#4CAF50"]

plt.pie(x, labels = my_labels, explode=my_explode, colors=my_colors)
plt.show() 
```

#### LEGEND
Use the `legend()` function to add an explanation for each wedge.  
Eg `plt.legend()`  
Use the `title` parameter in the `legend()` function to add a header to the legend.  
Eg `plt.legend(title="Animals")`  

```python
# LEGEND
import matplotlib.pyplot as plt
import numpy as np

x = np.array([20, 35, 15, 30])
my_labels = np.array(["Cats", "Dogs", "Ducks", "Geese"])
my_explode = [0.3,0,0.1,0]
my_colors = ["m", "crimson", "b", "#4CAF50"]

plt.pie(x, labels = my_labels, explode=my_explode, colors=my_colors)
plt.legend()
plt.show() 
```

```python
# LEGEND WITH HEADER
import matplotlib.pyplot as plt
import numpy as np

x = np.array([20, 35, 15, 30])
my_labels = np.array(["Cats", "Dogs", "Ducks", "Geese"])
my_explode = [0.3,0,0.1,0]
my_colors = ["m", "crimson", "b", "#4CAF50"]

plt.pie(x, labels = my_labels, explode=my_explode, colors=my_colors)
plt.legend(title="Animals")
plt.show() 
```
