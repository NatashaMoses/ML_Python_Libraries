#### CREATE SCATTER PLOTS
Scatter plots plot one dot for every observation.  
Use the `scatter()` function to create scatter plots.  
It is used in the **Pyplot** submodule.  
It takes two parameters:   
An array rep the **x-axis**   
An array rep the **y-axis**   
The no of points does not matter as long as there is same no in both x and y axes. 

#### CREATE MULTIPLE SCATTER PLOTS IN ONE FIGURE
Simply use the `scatter()` function twice to plot 2 scatter plots.  
The default plot colors are **blue** and **orange**.  
You can plot as many scatter plots as you like.  

```python
import matplotlib.pyplot as plt
import numpy as np

#week one, day and temperature:
x = np.array([0,1,2,3,4,5,6])
y = np.array([25,21,20,22,22,21,23])
plt.scatter(x, y)

#week two, day and temperature:
x = np.array([0,1,2,3,4,5,6])
y = np.array([20,20,21,23,24,22,21])
plt.scatter(x, y)

plt.show()
```

#### COLORS
+ Use the `color` or `c` argument to set the color of each scatter plot.   
E.g `plt.scatter(x, y, color = 'hotpink')`   
+ To set a specific color for each dot, use an array of colors as value for the `c` argument.  
E.g `colors = np.array(["red","green","blue","yellow","pink","black","orange"])` 
> NOTE  
> When setting a specific color for each dot, only use `c` argument not `colors`. 

```python
# COLOR FOR SCATTER PLOT
import matplotlib.pyplot as plt
import numpy as np

#week one, day and temperature:
x = np.array([0,1,2,3,4,5,6])
y = np.array([25,21,20,22,22,21,23])
plt.scatter(x, y, c = 'green')

#week two, day and temperature:
x = np.array([0,1,2,3,4,5,6])
y = np.array([20,20,21,23,24,22,21])
plt.scatter(x, y, c = 'purple')

plt.show()
```

```python
# COLOR FOR EVERY PLOT
import matplotlib.pyplot as plt
import numpy as np

#week two, day and temperature:
x = np.array([0,1,2,3,4,5,6])
y = np.array([20,20,21,23,24,22,21])

colors = np.array(["red","green","blue","yellow","pink","black","orange"])

plt.scatter(x, y, c = colors)

plt.show()
```

#### COLORMAP
This is a list of colors, where each color has a value that ranges from 0 to 100.  
There are many types of colormaps in **Matplotlib**.  
+ First create an array with values 0-100, one value for each point in the scatter plot.  
For example, if you have 6 points in the scatter plot, create an array with 6 points from 0-100.  
Eg `colors = np.array([0,20,40,60,80,100])`  
+ Use the keyword argument `cmap` to specify the value of the colormap.   
Eg `plt.scatter(x, y, c = colors, cmap = 'viridis')`  
+ Use `plt.colorbar()` to include the colormap in the figure.   

```python
# COLORMAP
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,1,2,3,4,5])
y = np.array([20,20,21,23,24,22])

colors = np.array([0,20,40,60,80,100])

plt.scatter(x, y, c = colors, cmap = 'viridis')

plt.colorbar()

plt.show()
```

#### SIZE
Use the `s` argument to change the size of the dots.  
+ First create an array for sizes that has same length as array for x and y axes.  
For example, if you have 6 points in the scatter plot, create an array with 6 points.  
Eg `sizes = np.array([20,50,120,35,60,75])`  
+ Use the keyword argument `s`.   
Eg `plt.scatter(x, y, s=sizes)`  

#### ALPHA
Use the `alpha` argument to adjust the transparency of the dots.    
Eg `plt.scatter(x, y, s=sizes, alpha=0.5)`  

```python
# SIZE OF THE DOTS
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,1,2,3,4,5])
y = np.array([20,20,21,23,24,22])

sizes = np.array([20,50,120,35,60,75])

plt.scatter(x, y, s = sizes)

plt.show()
```

```python
# SIZE AND ALPHA
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,1,2,3,4,5])
y = np.array([20,20,21,23,24,22])

sizes = np.array([20,50,120,35,60,75])

plt.scatter(x, y, s = sizes, alpha = 0.5)

plt.show()
```

```python
# COLORMAP WITH SIZE AND ALPHA
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,1,2,3,4,5])
y = np.array([20,20,21,23,24,22])

colors = np.array([0,20,40,60,80,100])
sizes = np.array([20,50,120,35,60,75])

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')

plt.colorbar()

plt.show()
```
