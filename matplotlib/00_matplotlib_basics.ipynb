#### PLOTTING  
The `plot()` function is used to draw points in graphs.  
It is used in the **Pyplot** submodule.  
It takes two parameters:   
An array rep the **x-axis**   
An array rep the **y-axis**   
The no of points does not matter as long as there is same no in both x and y axes.  
Use markers eg `'o'` to plot points only without a line.  
More about markers is in the next section.  
If only one axis is used, there are **default x-points** i.e 0,1,2,3...  

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1.2, 4, 5.4, 8])
y = np.array([4, 5, 7.8, 10.4])

plt.plot(x, y)
plt.show()
```

```python
# USING MARKER
plt.plot(x, y, 'o')
plt.show()
```

```python
# USING ONLY ONE AXIS (default x-axis becomes 0,1,2,3,4,..)
plt.plot(y, "o")
plt.show()
```

#### MARKERS  
The keyword `marker` is used to specify the type to marker.  
Eg `plt.plot(x, y, marker = 'D')`  
> NOTE  
> `plt.plot(x, y, 'D')` will plot the markers **without** lines.  
> `plt.plot(x, y, marker = 'D')` will plot the markers **with** lines.  
+ You can use shortcut string notation parameter.  
The parameter is **fmt**   
The syntax is `marker line color`  
Eg `plt.plot(x, y, 'H--m')`  
+ Use the keyword `markersize` or `ms` to set the size of the marker.  
Eg `plt.plot(x, y, 'H--m', ms = 20)`    
+ Use the keyword `markeredgecolor` or `mec` to set the color of the edge of the markers.  
Eg `plt.plot(x, y, 'H--m', ms = 20, mec = 'k')`  
+ Use the keyword `markerfacecolor` or `mfc` to set the color inside the edge of the markers.  
Eg `plt.plot(x, y, 'H--m', ms = 20, mec = 'k', mfc = 'y')`  
+ For color values, you can use hexadecimal values eg `mec = '#4CAF50'`  
+ You can also use any of the supported color values eg `mfc = 'crimson'`

```python
# USING ONLY ONE AXIS (default x-axis becomes 0,1,2,3,4,..)
plt.plot(y, marker = 'D')
plt.show()
```

```python
# USING fmt - marker line color
plt.plot(x, y, 'H--m')
plt.show()
```

```python
# USING ms, mec, mfc
plt.plot(x, y, 'H--m', ms = 20, mec = 'k', mfc = 'y')
plt.show()
```

#### LINES  
Use the keyword `linestyle` or `ls` to change the style of the plotted line.  
Eg `plt.plot(y, linestyle = 'dotted')`  
The following styles can be used:
|Style   | Or            |
|-----:  |---------------|
|'-'     |'solid'        |
|':'     |'dotted'       |
|'--'    |'dashed'       |
|'-.'    |'dashdot'      |
|"" or ''|'None'         |
+ Use the keyword `color` or `c` to set the color of the line.  
Eg `plt.plot(x, y, color = 'b')`  
You can also use hexadecimal or any of the supported color values.  
+ Use the keyword `linewidth` or `lw` to change the width of the line.  
Eg `plt.plot(x, y, linewidth = '18.5')`  
+ To plot many lines, use more `plt.plot()` functions.  


```python
# LINESTYLE, LINECOLOR AND LINEWIDTH
plt. plot(x, y, ls = ':', c = 'crimson', lw = '2.5')
plt.show()
```

```python
# PLOT MULTIPLE LINES
y1 = np.array([2, 5, 7, 9.5])
y2 = np.array([3.5, 5.2, 8, 10])

plt.plot(x, y1, 'o--m')
plt.plot(x, y2, '*:y')

plt.show()
```

#### LABELS  
+ Use the `xlabel()` function to set a label for the x-axis.  
+ Use the `ylabel()` function to set a label for the y-axis.  
+ Use the `title()` function to set a title for the plot.  
+ Use the `fontdict` parameter to set the font properties for labels/titles.  
+ Use the `loc` parameter to position the title.  

```python
# PLOT WITH TITLE AND X/Y LABELS
weight = np.array([55, 68, 73, 90])
height = np.array([5.5, 5.8, 6.1, 6.5])

plt.title("Weight vs Height")
plt.xlabel("Height")
plt.ylabel("Weight")

plt.plot(weight, height, 'o--m')
plt.show()
```

```python
# APPLY FONTS TO TITLE AND X/Y LABELS
# POSITION THE TITLE
weight = np.array([55, 68, 73, 90])
height = np.array([5.5, 5.8, 6.1, 6.5])

font1 = {'family':'serif','color':'hotpink','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Weight vs Height", fontdict = font1, loc = 'left')
plt.xlabel("Height", fontdict = font2)
plt.ylabel("Weight", fontdict = font2)

plt.plot(weight, height, 'o--m')
plt.show()
```

#### GRID  
Use the `grid()` function to add grid lines to the plot.  
This function is used in the **Pyplot** submodule.  
+ Use the `axis` parameter in the `grid()` function to specify which grid lines to display.  
The accepted values are:  
 `'x'` Display grid lines for x-axis only.    
 `'y'` Display grid lines for y-axis only.  
 `'both'` This is the default, display both x and y axes.    
+ Use the keyword `linestyle` or `ls` to change the style of the grid lines.  
Eg `plt.grid(linestyle = 'dotted')`  
+ Use the keyword `color` or `c` to set the color of the grid lines.  
Eg `plt.grid(color = 'b')`  
You can also use hexadecimal or any of the supported color values.  
+ Use the keyword `linewidth` or `lw` to change the width of the grid lines.  
Eg `plt.grid(linewidth = '18.5')`  


```python
# DISPLAY GRID LINES
weight = np.array([55, 68, 73, 90])
height = np.array([5.5, 5.8, 6.1, 6.5])

plt.title("Weight vs Height")
plt.xlabel("Height")
plt.ylabel("Weight")

plt.grid()
plt.plot(weight, height, 'o--m')
plt.show()
```

```python
# DISPLAY GRID LINES FOR X-AXIS ONLY
# USE color, linestyle and linewidth
weight = np.array([55, 68, 73, 90])
height = np.array([5.5, 5.8, 6.1, 6.5])

plt.title("Weight vs Height")
plt.xlabel("Height")
plt.ylabel("Weight")

plt.grid(axis = 'x', c = 'aqua', ls = '--', lw = 0.7 )
plt.plot(weight, height, 'o--m')
plt.show()
```

#### SUBPLOT
The `subplot()` function can be used to draw multiple plots in one figure.  
It takes three arguments;
+ The **first** argument represents the **rows**.  
+ The **second** argument represents the **columns**.  
+ The **third** argument represents the **index of the current plot**.   
A plot with 1 row and 2 columns means the plots are displayed side by side.  
A plot with 2 rows and 1 colimn means the plots are displayed on top of each other.  
However, you can have as many rows and columns.  
+ Use the `suptitle()` function to add one title for all plots.
